```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#이미지 크기 지정
img_width = 250
img_height = 250

#train과 test 비율 설정
test_ratio = 3/4
```

```python
tfrecord_train = 'train.tfrecord'
tfrecord_test = 'test.tfrecord'
tfrecord_dir = 'tfrecords'

if not os.path.exists(tfrecord_dir):
    print('hello')
    os.makedirs(tfrecord_dir)

normal_image_dir = 'lfw_funneled'
blurred_image_dir = 'lfw_blurred'
normal_image_path = os.path.join(os.getcwd(), normal_image_dir)
blurred_image_path = os.path.join(os.getcwd(), blurred_image_dir)
```

```python
normal = []
blurred = []
label_normal = []
label_blurred = []
```

```python
for folder in os.listdir(normal_image_path):
    print(folder)
    full_path = os.path.join(normal_image_path, folder)
    for file in os.listdir(full_path):
        print(file)
        image = Image.open(os.path.join(full_path, file))
        image.resize((img_width, img_height))
        image = np.asarray(image)
        #이미지 array를 리스트에 넣는다.
        normal.append(image)
        label_normal.append(0)
```

```python
normal = np.asarray(normal)
print(normal.shape)
```

```python
for folder in os.listdir(blurred_image_dir):
    print(folder)
    full_path = os.path.join(blurred_image_dir, folder)
    for file in os.listdir(full_path):
        print(file)
        image = Image.open(os.path.join(full_path, file))
        image.resize((img_width, img_height))
        image = np.asarray(image)
        blurred.append(image) 
        label_blurred.append(1)
```

```python
blurred = np.asarray(blurred)
pic = np.vstack((normal, blurred))

label = label_normal + label_blurred
label = np.asarray(label)
print(pic.shape)
print(label.shape)
```

```python
len_pic = len(pic)
idxrand = np.random.permutation(len_pic)
idxtrain = idxrand[:int(test_ratio*len_pic)]
idxtest = idxrand[int(test_ratio*len_pic):]
print(len(idxtrain), len(idxtest))
```

```python
## tfrecord file 경로 설정
train_tfr_path = os.path.join(os.getcwd(), tfrecord_dir, tfrecord_train)
test_tfr_path = os.path.join(os.getcwd(), tfrecord_dir, tfrecord_test)
print(train_tfr_path)
```

```python
## tfrecord writer
writer_train = tf.python_io.TFRecordWriter(train_tfr_path) 
writer_test = tf.python_io.TFRecordWriter(test_tfr_path)
```

```python
def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value])) #Integer 용 Feature 저장

def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value])) #String, byte용 Feature 저장 json과 같은 포맷으로 저장이 된다.
```

```python
import cv2
```

```python
train_cnt = 0
for idx in idxtrain:
    temp_image = cv2.imencode('.jpg', pic[idx])[1].tostring()
    image_str = tf.compat.as_bytes(temp_image)
    
    #image_str = pic[idx].tobytes()
    label_one = label[idx]    
    
    example = tf.train.Example(features=tf.train.Features(feature={
        'image':_bytes_feature(image_str),
        'label':_int64_feature(label_one)
    }))
    writer_train.write(example.SerializeToString())
    train_cnt += 1
    
print ("{} train images are saved.".format(train_cnt))  
```

```python
test_cnt = 0
for idx in idxtest:
    temp_image = cv2.imencode('.jpg', pic[idx])[1].tostring()
    image_str = tf.compat.as_bytes(temp_image)
    
    label_one = label[idx]
    
    example = tf.train.Example(features=tf.train.Features(feature={
        'image':_bytes_feature(image_str),
        'label':_int64_feature(label_one)
    }))
    writer_test.write(example.SerializeToString())
    test_cnt += 1
print ("{} train images are saved.".format(train_cnt))  
```

```python
print ("{} train images are saved.".format(test_cnt))  
```

```python
writer_train.close()
writer_test.close()
```
